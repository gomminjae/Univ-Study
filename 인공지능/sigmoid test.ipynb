```python
import tensorflow as tf
import numpy as np
```

```python
data = np.array([[1,1], [4,2], [3,2], [7,3]], dtype = np.float32)
x = data[:,0:1]
y = data[:,1:]

px = tf.placeholder(tf.float32 , shape = [None, 1])
py = tf.placeholder(tf.float32 , shape = [None,1])

w1 = tf.Variable(tf.random_normal([1,1]), name = "weight")
x2 = tf.sigmoid(tf.matmul(px, w1)) #activation 

w2 = tf.Variable(tf.random_normal([1,1]), name = "weight")
x3 = tf.sigmoid(tf.matmul(x2, w2))

w3 = w1 = tf.Variable(tf.random_normal([1,1]), name = "weight")

h = tf.matmul(x3, w3)

loss = tf.reduce_mean(tf.square(h - py))
opti = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = opti.minimize(loss)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for i in range(1000):
    feed = {px:x, py:y}
    sess.run(train, feed_dict = feed)
    print(i, sess.run(loss, feed_dict = feed), sess.run(w3))
```
