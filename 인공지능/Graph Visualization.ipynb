## Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```


## Basic Attributes
- alpha: 투명도
- kind: 그래프 종류

## 점선 그래프

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

## 여러 그래프 

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.show()
```

## Multi Graph

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2,2,1)
plt.plot(data)

plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)

plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3)

plt.show()
```

## 그래프 선 옵션 
- 색상: rgb, c(청록색), y(노란색), k(검은색), w(흰색)
- 마커: o(원), v(역삼각형), ^(삼각형), s(사각형), +(플러스), .(점)

```python
plt.plot(data, 'y+')
plt.show()
```

## 그래프 사이즈 조절
- plt.figure안의 figsize로 가로 세로 조절

```python
plt.figure(figsize=(10,6))
plt.plot(data, 'k+')
plt.show()
```

## 그래프 겹치기 & legend 달기

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label = 'step_post')
plt.legend()
plt.show()
```

## 이름 달기

```python
plt.plot(np.random.randn(1000).cumsum())
plt.title("Random Graph")
plt.xlabel('Stages')
plt.ylabel('Values')
plt.show()
```

## 종합

```python
plt.plot(np.random.randn(1000).cumsum(), 'k', label='one')
plt.plot(np.random.randn(1000).cumsum(), 'b', label='two')
plt.plot(np.random.randn(1000).cumsum(), 'r', label='three')

plt.legend()
plt.show()
```
